```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:\\data science\\assignment\\sales_data_with_discounts (2).csv")
data
```

```python
numerical_data = data.select_dtypes(include=['number'])
```

```python
#mean 
mean = numerical_data.mean()
print("Mean:\n", mean)
```

```python
# Median
median = numerical_data.median()
print("\nMedian:\n", median)
```

```python
# Mode
mode = numerical_data.mode().iloc[0]
print("\nMode:\n", mode)
```

```python
# Standard Deviation
std_dev = numerical_data.std()
print("\nStandard Deviation:\n", std_dev)
```

```python
#brief interpretation 
#Mean: Average value of the data.

#Median: Middle value of the data.

#Mode: Most frequent value.

#Standard Deviation: Shows how much the data varies from the mean.
```

```python
plt.hist(data['Net Sales Value'], bins=10)
plt.xlabel("Net Sales Value")
plt.ylabel("Frequency")
plt.title("Histogram of Net Sales Value")
plt.show()
```

```python
plt.figure()
data.boxplot(column='Net Sales Value', by='City')
plt.xlabel("City")
plt.ylabel("Net Sales Value")
plt.title("Net Sales Value by City")
plt.show()
```

```python
brand_counts = data['Brand'].value_counts()

plt.bar(brand_counts.index, brand_counts.values)
plt.xlabel("Brand")
plt.ylabel("Count")
plt.title("Bar Chart of Brand Distribution")
plt.show()
```

```python
#The key findings from the descriptive analytics and data visualizations.
#istogram of Net Sales Value shows a right-skewed distribution, where most sales values are concentrated at the lower range, with a few very high sales values extending the tail.

#This skewness indicates the presence of high-value transactions, which increase the mean compared to the median.

#The box plot of Net Sales Value by City reveals a wide spread in sales values and the presence of multiple outliers, suggesting significant variation in sales performance.

#The median sales value is relatively low compared to the maximum values, confirming uneven sales distribution.

#The bar chart of Brand distribution shows that some brands (such as Jeera) appear much more frequently than others, indicating brand dominance in the dataset.
```
